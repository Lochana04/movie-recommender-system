```python
# 🎬 Smart Movie Recommender System

#This project recommends movies based on content similarity using NLP and Machine Learning techniques.
```

```python
#importing Libraries
```

```python
import pandas as pd
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Load dataset
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

movies.head()
```

```python
#Merging Dataset
```

```python
movies = movies.merge(credits, on='title')
movies.head()
```

```python
#select important columns
```

```python
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
#Rename column
```

```python
movies = movies.rename(columns={'id': 'movie_id'})
```

```python
#check missing values
```

```python
movies.isnull().sum()
```

```python
#handle missing values
```

```python
movies.dropna(inplace=True)
```

```python
#generate json columns
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#extracting cast
```

```python
def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
#extrating Director
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#Text Processing
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
#Remove spaces
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
#creating tags
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
#Final dataset
```

```python
new_df = movies[['movie_id','title','tags']]

new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())

new_df.head()
```

```python
#new_df.to_csv("processed_movies.csv", index=False)
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#Vectorization - Feature Engineering
```

```python
similarity = cosine_similarity(vectors)
```

```python
#Recommend Function
```

```python
def recommend(movie):
    if movie not in new_df['title'].values:
        print("Movie not found!")
        return

    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
#Testing
```

```python
recommend("Avatar")
```

```python
#saving processed dataset
```

```python
new_df.to_csv("processed_movies.csv", index=False)
```
